```python
#imported necesarry libraries for graphing
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

#Data points for one of our sets 
x = ([27.4,25.52,25.27,26.89,24.66,24.4,27.44,27.53,27.55,27.76,20.85,20.06,20.2,20.15])
y  =([16, 12,8.9,14.1,10.4,14,22.2,23.9,19,85.7,1.3,0,47.4,10.4])

#Plot the data as a scatter plot
plt.plot(x,y,'ro')
#Example of title shown
plt.title(" Temperature vs. Turbidity ")
plt.xlabel("Temperature(C)")
plt.ylabel("Turbidity(NTU)")
plt.show()

#Array of data points for our set
x = np.array([27.4,25.52,25.27,26.89,24.66,24.4,27.44,27.53,27.55,27.76,20.85,20.06,20.2,20.15])
y = np.array([16, 12,8.9,14.1,10.4,14,22.2,23.9,19,85.7,1.3,0,47.4,10.4])

linreg = LinearRegression()
x = x.reshape(-1,1)
linreg.fit(x,y)

#Predicted value for y that follows the linear trend
y_pred = linreg.predict(x)
plt.scatter(x,y)
#Plots the data points using the y predicted and x values we used
plt.plot(x,y_pred,color = "green")
plt.show()

#Slope for linear regression
print(linreg.coef_)
#Y-Intercept for linear regression
print(linreg.intercept_)

```
