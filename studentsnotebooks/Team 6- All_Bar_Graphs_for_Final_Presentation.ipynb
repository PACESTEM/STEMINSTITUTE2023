```python

import pandas as pd
from matplotlib import pyplot as plt
 
# Read CSV into pandas
data = pd.read_csv(r"bop_data2.csv.txt")
data.head()
df = pd.DataFrame(data)
 
X = df['shell_height_mm']
Y = df['Site']
 
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(Y, X, color = "blue")
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
plt.ylabel("Site")
plt.xlabel("shell_height_mm")
# Show Plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
'''Have the y axis be the sites Requirement meet
and the x axis be the mean shell height
for each of the sites
'''
# Read CSV into pandas
data = pd.read_csv(r"bop_data2.csv.txt")

data.head()
df = pd.DataFrame(data)
 
oysterMean = df[["Site","shell_height_mm"]].groupby("Site").mean()

oysterMeanValues = []

for i in range(0,9):
    oysterMeanValues.append(oysterMean.iloc[i][0])

oysterSites = ["Bush Terminal Park", "Brooklyn Bridge Park","Coneny Island Creek","Lemon Creek Lagoon","Paerdegat Basin","Governor's Island","Lemon Creek Nursey","Great Kills Harbor","Brooklyn Navy Yard"]
oysterSites.sort()

X = oysterMeanValues
Y = oysterSites
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(Y, X)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
plt.ylabel("Site Name")
plt.xlabel("Mean Shell Height In Millimeters")
# Show Plot
plt.show()
```
