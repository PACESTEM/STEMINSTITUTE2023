```python
# Oyster data with water color data from 2018 
measures = pd.read_csv(url + 'bop_data.csv')
```

```python
# we can calculate the correlation in several ways as well
# np.corrcoef returns an np array
measures['pH'] = measures['pH'] + np.random.normal(0,4,measures.shape[0])
np.corrcoef(measures['pH'], measures['pH'])
```

```python
corr = measures.corr()
```

```python
fig, ax = plt.subplots(figsize=(20,10)) 
sns.heatmap(measures.corr(), annot = True, fmt = ".2f", linewidths=0.5, cmap='viridis', ax=ax) 
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
